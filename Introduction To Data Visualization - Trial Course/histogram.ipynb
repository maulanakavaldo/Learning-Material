## Bar Chart

Bar Chart merupakan bentuk visualisasi data, dengan komponen dasar:
* Memiliki dua sumbu
* Satu sumbu yang menunjukkan variasi kategori.
* Sumbu lainnya menunjukkan frekuensi kejadian atau nilai dari variabel.


Contoh, Bar Chart menunjukkan warna favorit anak-anak:
<center><img src="https://images.twinkl.co.uk/tw1n/image/private/t_630/u/ux/barchart_ver_1.jpg" width="300"/></center>

### Let's code

Misal, diberikan data mengenai mata pelajaran favorit dari siswa pada sebuah kelas:

| Subject    | N Students |
|------------|------|
| Maths      | 2    |
| English    | 4    |
| Arts       | 5    |
| History    | 3    |
| PE         | 5    |
| Music      | 4    |
| Technology | 5    |

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
subjects = ["Maths", "Eng", "Arts", "Hist", "PE", "Music", "Tech"]
n_stud   = [2,        4,     5,      3,      5,    4,       5    ]
```

```python
plt.bar(subjects, n_stud)
plt.title("Favorite Subject")
plt.show()
```

Ukuran plot dapat kita sesuaikan dengan fungsi `figure` dengan parameter `figuresize`

```python
plt.figure(figsize=(10,5))
plt.bar(subjects, n_stud)
plt.title("Favorite Subject")
plt.show()
```

Warna dari plot juga dapat diubah dengan menggunakan argumen `color`

```python
plt.figure(figsize=(10,5))
plt.bar(subjects, n_stud, color='r')
plt.title("Favorite Subject")
plt.show()
```

## Grouped Bar Chart

Grouped Bar Chart merupakan bentuk visualisasi data, dengan komponen dasar:

* Memiliki dua sumbu
* Satu sumbu yang menunjukkan variasi grup.
* Sumbu lainnya menunjukkan frekuensi kejadian atau nilai dari variabel.
* Memiliki kode warna untuk menunjukkan variasi kategori. 

Contoh, Grouped Bar Chart menunjukkan warna favorit anak-anak:
<center><img src="https://images.twinkl.co.uk/tw1n/image/private/t_630/u/ux/grouped-bar-chart-new_ver_1.jpg" width="300"/></center>

### Let's code

Misal, diberikan data mengenai warna favorit dari siswa pada tiap-tiap tingkatan pendidikan:

| school     |Green |Blue  |Yellow|red   |
|------------|------|------|------|------|
| Preschool  | 4    | 2    | 5    | 4    |
| Primary    | 2    | 4    | 6    | 3    |
| Secondary  | 1    | 6    | 4    | 4    |


```python
school = ['Preschool', 'Primary', 'Secondary']
green = [4,2,1]
blue = [2,4,6]
yellow = [5,6,4]
red = [4,3,4]
```

Cara paling mudah untuk membuat grouped bar chart adalah dengan menggunakan bantuan `pandas`

```python
import pandas as pd
```

```python
df = pd.DataFrame([school, green, blue, yellow,red]).T
df.columns = ['school','green','blue','yellow','red']
```

```python
df
```

```python
my_colors = ['g', 'b', 'y', 'r']
ax = df.plot(x='school',
        kind='bar',
        stacked=False,
        color = my_colors,
        rot = 0,
        title='Favorite Color of Students')
ax.legend(bbox_to_anchor=(1.0, 1.0))
plt.show()
```

## Histogram

* Histogram digunakan untuk menunjukkan perkiraan represetasi dari distribusi data numerik.

* Histogram menggunakan binning(rentang data) dengan ukuran yang sama. Tinggi dari bar dalam histogram menunjukkan frekuensi dari data yang sudah dikelompokkan ke dalam tiap bin.

* Umumnya, histogram digunakan pada data numerik kontinu untuk mengetahui frekuensi pada range tertentu.

* Untuk menggambarkan histogram, gunakan `plt.hist()`.

Contoh: Histogram menunjukkan umur perempuan yang mengunjungi perpustakaan
<center><img src="https://www.softschools.com/math/probability_and_statistics/images/histogram_2.png" width="400"/></center>

### Let's Code

Misal, diberikan data berupa umur orang pada suatu desa dengan jumlah penduduk 700 orang:

```python
# Untuk saat ini, abaikan kode dibawah
# Kode hanya digunakan untuk membuat sampel

age = np.random.gamma(4, 6, 700).astype(int)
```

```python
age[:10]
```

Membuat histogram dari umur penduduk

```python
plt.hist(age)
plt.xlabel('Umur')
plt.ylabel('Jumlah Penduduk')
plt.show()
```

`seaborn`

```python
import seaborn as sns
```

```python
sns.histplot(age)
plt.xlabel('Umur')
plt.ylabel('Jumlah Penduduk')
plt.show()
```

Namun misalkan pada histogram tersebut, rentang tiap bar masih dirasa kurang tepat, kita dapat memodifikasi ukuran dari rentang (binning) dari histogram.

```python
plt.hist(age, bins=25)
plt.xlabel('Umur')
plt.ylabel('Jumlah Penduduk')
plt.show()
```

```python
sns.histplot(age, bins=25)
plt.xlabel('Umur')
plt.ylabel('Jumlah Penduduk')
plt.show()
```
