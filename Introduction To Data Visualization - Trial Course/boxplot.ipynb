```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

# Apa itu Boxplot?

Box and whisker plot adalah grafik 2 dimensi yang mana menggunakan box dan garis untuk menggambarkan distribusi dari satu atau lebih kumpulan data numerik.


```python
plt.figure(figsize = [10, 10])
sns.boxplot(
    x = sns.load_dataset("flights")["year"],
    y = sns.load_dataset("flights")["passengers"]
)
```

Boxplot umum digunakan saat kita mempunyai satu atau banyak kelompok data dan kita ingin melihat distribusinya bersama-sama, contoh use-cases:
- Melihat distribusi dari expenses kita di suatu bulan atau di setiap bulan selama tahun 2021.
- Cek distribusi sales perusahaan dari x range tahun.
- dll.

# Bagaimana membuatnya?

Untuk membuat boxplot ada banyak cara, bisa menggunakan fungsi **boxplot** yang telah disediakan oleh matplotlib atau bisa juga menggunakan fungsi **boxplot** dari seaborn.

Sekilas mengenai **Seaborn** dan **Matplotlib**: <br>
**Seaborn** adalah library visualisasi data selayaknya **Matplotlib**, **Seaborn** dibangun berdasarkan **Matplotlib** sehingga memiliki kelebihan yaitu lebih *friendly* untuk digunakan. <br>

Sintaks membuat boxplot pada seaborn adalah sebagai berikut: <br>
> `sns.boxplot(x = data_x_axis, y = data_y_axis)` <br>
<br>
> **x**: berupa data kategorikal maupun numerikal yang akan mengelompokkan data pada *y axis*. <br>
> **y**: berupa data numerkal baik diskrit maupun kontinu yang dikelompokkan oleh data pada *x axis*. Tidak bisa data kategorikal karena tidak bisa dicari informasi statistiknya seperti rata-rata, nilai tengah maupun nilai yang sering muncul.<br>
> <br>

# Interpretasi dari Boxplot

```python
dataset = sns.load_dataset("flights").copy()
dataset = pd.concat([dataset, pd.DataFrame({"year": 1960, "month": "Jan", "passengers": 1000}, index = [len(dataset)])])
dataset = pd.concat([dataset, pd.DataFrame({"year": 1960, "month": "Dec", "passengers": 250}, index = [len(dataset)])])

plt.figure(figsize = [10, 10])
sns.boxplot(
    x = dataset[dataset["year"] == 1960]["year"],
    y = dataset[dataset["year"] == 1960]["passengers"]
)
```

Secara garis besar **boxplot** dibagi menjadi 2 bagian:
1. Box biru.
2. Garis pada bagian atas dan bawahnya.

Karena orientasi **boxplot** ini vertikal maka menggunakan kata-kata atas dan bawah, jika orientasinya horizontal maka menggunakan kata-kata kiri dan kanan. <br>
<br>
Bagian-bagian dari **boxplot**! Dimulai dari bawah.<br>
1. Titik pada bagian paling bawah, dibawah nilai **passengers** 300 merupakan outlier.
2. Garis horizontal paling bawah yang letaknya dibawah nilai **passengers** 400 hingga menyentuh box biru merupakan bagian minimum dari garis **whisker**, bagian ini merupakan observasi paling bawah sebelum batas bawah (Q1 - (1.5 x IQR)).
3. Garis yang menghubungkan nilai minimum pada nomor 1 dengan box biru bernama **whisker**, garis ini menunjukkan bahwa data pada bagian tersebut bernilai lebih rendah dari nilai IQR.
4. Box biru, box biru sendiri dibagi menjadi 3 bagian yaitu:
     1. **Q1** atau **kuartil pertama** yang letaknya berada pada bagian bawah dari box biru atau yang langsung bersentuhan dengan garis **whisker**.
     2. **Q2** atau **kuartil kedua** atau **median** merupakan garis ditengah box biru.
     3. **Q3** atau **kuartil ketiga** yang letaknya berada pada bagian atas dari box biru atau yang langsung bersentuhan dengan garis **whisker** bagian atas.
5. Garis **whisker** atas.
6. Garis horizontal paling atas adalah bagian dari garis **whisker** yang merupakan observasi paling atas sebelum batas atas (Q3 + (1.5 x IQR))
7. Titik paling atas merupakan outlier.

# Kostumisasi Boxplot

- Mengganti orientasi **boxplot**.
- Mengganti warna pada **boxplot**.
- Resize plot.

### 1. Mengganti orientasi **boxplot**.

Mengganti orientasi pada boxplot juga berhubungan dengan *x-axis* dan *y-axis*.
- **y-axis**: year.

```python
plt.figure(figsize = [10, 10])
sns.boxplot(
    x = sns.load_dataset("flights")["passengers"],
    y = sns.load_dataset("flights")["year"]
)
```

Tidak jelas bukan?
Sekarang mengganti orientasinya dengan menggunakan parameter:
> `orient = "v" atau "h"`

```python
plt.figure(figsize = [10, 10])
sns.boxplot(
    x = sns.load_dataset("flights")["passengers"],
    y = sns.load_dataset("flights")["year"],
    orient = "h"
)
```

### 2. Mengganti warna boxplot.

Mengganti warna dari **boxplot** ada 2 tipe:
1. Hanya jadi 1 warna.
2. Tiap **boxplot** mempunyai warna berbeda
<br>

Jika ingin ganti menjadi 1 warna saja, maka perlu buka website list of named colors dari **Matplotlib**: <br>
https://matplotlib.org/stable/gallery/color/named_colors.html <br>
<br>
Jika ingin mengganti warna namun tetap berbeda untuk tiap **boxplot**, maka dibutuhkan nama pallete yang mana terdapat pada website matplotlib: <br>
https://matplotlib.org/stable/tutorials/colors/colormaps.html. <br>
<br>
tambahkan parameter:
> `color = 'named_color'`. <br>
> atau <br>
> palette = 'palette_name'. <br>
<br>

Pada contoh kali ini menggunakan nama color = **tab:blue** dan nama palette **Pastel1**. <br>
Nama tersebut adalah **CASE SENSITIVE**.

```python
plt.figure(figsize = [10, 10])
sns.boxplot(
    x = sns.load_dataset("flights")["passengers"],
    y = sns.load_dataset("flights")["year"],
    orient = "h",
    color = "tab:blue"
)
```

jika ingin warna yang berbeda untuk tiap **boxplot**nya maka menggunakan parameter:
> `palette = nama_palette`

```python
plt.figure(figsize = [10, 10])
sns.boxplot(
    x = sns.load_dataset("flights")["passengers"],
    y = sns.load_dataset("flights")["year"],
    orient = "h",
    palette = "Pastel1"
)
```

### 3. Resize boxplot.

Untuk resize boxplot hanya perlu menambahkan 1 line of code yaitu

> `plt.figure(figsize([horizontal_size, vertical_size]))`

Contoh kali ini menggunakan ukuran **20 x 20**. <br>
Kode diatas **harus** dijalankan terlebih dahulu sebelum membuat **boxplot**.


```python
plt.figure(figsize = [20, 20])
sns.boxplot(
    x = sns.load_dataset("flights")["passengers"],
    y = sns.load_dataset("flights")["year"],
    orient = "h",
    palette = "Pastel1"
)
```
